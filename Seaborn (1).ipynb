```python
import seaborn
seaborn.set(style = 'whitegrid')
data = pandas.read_csv( "data.csv" )
seaborn.swarmplot(x = data["tip"],y=data['sex'])
```

```python
import seaborn
seaborn.set(style = 'whitegrid')
data = pandas.read_csv("data.csv")
seaborn.pointplot(x = "size", y = "tip", data = data)
```

#### Seaborn Lineplot is a data visualization tool that depicts the relationship between a set of continuous and categorical data points. 
#### A scatter plot is used to obtain the correlational relationship between two numerical variables. 
#### A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.
#### swarmplot() method is used to draw a non-overlapping scatter plot where one of the variables is a categorical variable.
#### A point plot represents an estimate of central tendency for a numeric variable by the position of the dot and provides some indication of the uncertainty around that estimate using error bars
